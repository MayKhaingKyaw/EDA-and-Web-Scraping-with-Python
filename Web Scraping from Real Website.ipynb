```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.iban.com/country-codes"
```

```python
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
table=soup.find('table',class_='table table-bordered downloads tablesorter')
```

```python
print(table)
```

```python
country_codes=table.find_all('th')
```

```python
country_codes
```

```python
country_codes_list=[title.text for title in country_codes]
print(country_codes_list)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns =country_codes_list)
```

```python
df
```

```python
column_data=table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data=row.find_all('td')
    individual_row_data=[data.text.strip() for data in row_data]
    #print(individual_row_data)
    length=len(df)
    df.loc[length]=individual_row_data
```
